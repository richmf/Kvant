# Líneas de campo eléctrico producidas por  cargas

> Dr. Ricardo Méndez Fragoso $\mathbb{R}i \vec c \hbar$

El campo eléctrico, $\vec E$, que produce una carga o varias cargas esta dada por la siguiente expresión:
$$
\vec E = \frac{1}{4 \pi \epsilon _0} \sum _{i=1} ^N q_i \frac{\vec r - \vec r_{q_i}}{|\vec r - \vec r_{q_i}|^3}
$$

```python
from pylab import *
from scipy.integrate import *
```

```python
def carga(x,x_q,q):
    r_rq = ((x[0]-x_q[0])**2 + (x[1]-x_q[1])**2 )**1.5
    Ex = (x[0]-x_q[0])/r_rq
    Ey = (x[1]-x_q[1])/r_rq
    return array([q*Ex,q*Ey])
```

```python
x,y = linspace(-3,3),linspace(-3,3)
xv,yv = meshgrid(x,y)
E = carga([xv,yv],[0,1],1) + carga([xv,yv],[0,-1],-1)
figure(dpi=200)
streamplot(xv,yv,E[0],E[1],density=2,linewidth=1)
show()
```

```python
x,y = linspace(-3,3),linspace(-3,3)
xv,yv = meshgrid(x,y)
E = carga([xv,yv],[0,1],-1) + carga([xv,yv],[0,-1],-1)
E += carga([xv,yv],[1,0],1) + carga([xv,yv],[-1,0],1)
figure(dpi=200)
streamplot(xv,yv,E[0],E[1],density=2,linewidth=1)
plot([0,0],[1,-1],'o',color="tab:red")
plot([1,-1],[0,0],'o',color="tab:green")
show()
```

```python
x,y = linspace(-3,3),linspace(-3,3)
xv,yv = meshgrid(x,y)
E = carga([xv,yv],[0,1],-1) + carga([xv,yv],[0,-1],-1)
E += carga([xv,yv],[1,0],1) + carga([xv,yv],[-1,0],1)
E += carga([xv,yv],[-2,0],1) + carga([xv,yv],[2,0],1)
figure(dpi=200)
streamplot(xv,yv,E[0],E[1],density=2,linewidth=1)
plot([0,0],[1,-1],'o',color="tab:red")
plot([1,-1,-2,2],[0,0,0,0],'o',color="tab:green")
show()
```
